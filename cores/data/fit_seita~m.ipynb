# 拟合 θ~m 曲线

```python
from scipy.optimize import leastsq
import numpy  as np
import matplotlib.pyplot as plt

def fit(file):
    with open(file, 'r') as fp:
        info = [[float(y) for y in x.strip().split(',')] for x in fp.read().split('\n') if len(x) < 15 and len(x) > 1]
    seitas, ms = zip(*info)
    
    def f(args, seitas, ms):
        a, b  = args
        seitas = np.array(seitas)
        ms = np.array(ms)
        return a*seitas**b - ms

    result = leastsq(f, [0.1, 0.1], args=(seitas, ms))

    def ff(seita):
        a, b = result[0]
        return a*seita**b

    x = np.linspace(5, 100, 1000)
    y = ff(x)
    
    print(list(zip(seitas, ms)))
    print('m = %f*θ^%f' % tuple(result[0]))
    plt.figure(figsize=(18, 12))
    plt.plot(y, x, c='r')
    plt.plot(ms, seitas, c='b')
    plt.show()
```

```python
fit('Seita-m_area1.dat')
```
